```python
import numpy as np
```

```python
text = ( """⭐ Swami Vivekananda (1863–1902)

🔹 Great Indian philosopher and spiritual leader.
🔹 Disciple of Ramakrishna Paramahamsa.
🔹 Introduced the teachings of Vedanta and Yoga to the Western world.
🔹 Famous for his speech at the Parliament of the World's Religions in Chicago in 1893.
🔹 Promoted self-confidence, unity, and service to humanity.
🔹 Founded the Ramakrishna Mission in 1897.
🔹 Inspired millions of people through his teachings and writings.
🔹 His birthday, 12 January, is celebrated as National Youth Day in India. 🇮🇳

✨ Quote: "Arise, awake, and stop not till the goal is reached.""" )
```

```python
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
tokens= word_tokenize(text)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.isalpha() and token.lower() not in stop_words]
unique_tokens_count = len(set(filtered_tokens))
print("Number of unique tokens:", unique_tokens_count)
unique_tokens = set(filtered_tokens)
print("Unique Tokens:", unique_tokens)
```

```python
import string
```

```python
punctuation_count = sum(1 for char in text if char in string.punctuation)
text_without_punctuation = ''.join(char for char in text if char not in string.punctuation)
print("Number of punctuation characters:", punctuation_count)
print("Text without punctuation:", text_without_punctuation)
```

```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags = nltk.pos_tag(tokens)
pos_tags_count = Counter(tag for _, tag in pos_tags)
```

```python
plt.bar(pos_tags_count.keys(), pos_tags_count.values())
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags in Text')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from nltk import bigrams, trigrams, ngrams
```

```python
bigrams_list = list(bigrams(tokens))
trigrams_list = list(trigrams(tokens))
ngrams_list = list(ngrams(tokens, 5))
```

```python
unique_bigrams = set(bigrams_list)
unique_trigrams = set(trigrams_list)
unique_ngrams = set(ngrams_list)
```

```python
print(len(unique_bigrams))
print(len(unique_trigrams))
print(len(unique_ngrams))
```
